```python
with open('/content/drive/MyDrive/moby_dick.txt') as story:
  story_data = story.read()

print(story_data)
```

```python

import tensorflow as tf
import pandas as pd
import numpy as np
import re
```

```python
def clean_text(text):
  text = re.sub(r',', '', text)
  text = re.sub(r'\'', '',  text)
  text = re.sub(r'\"', '', text)
  text = re.sub(r'\(', '', text)
  text = re.sub(r'\)', '', text)
  text = re.sub(r'\n', '', text)
  text = re.sub(r'', '', text)
  text = re.sub(r'', '', text)
  text = re.sub(r'', '', text)
  text = re.sub(r'\.', '', text)
  text = re.sub(r';', '', text)
  text = re.sub(r':', '', text)
  text = re.sub(r'\-', '', text)

  return text
```

```python
lower_data = story_data.lower()
```

```python
split_data = lower_data.splitlines()
print(split_data)

final = ''


print(final)
```

```python
for line in split_data:
  line = clean_text(line)
  final += '\n' + line

print(final)
```

```python
final_data = final.split('\n')
print(final_data)
```

```python
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences
```

```python
max_vocab = 1000000
tokenizer = Tokenizer(num_words=max_vocab)
tokenizer.fit_on_texts(final_data)
```

```python
word2idx = tokenizer.word_index
print(len(word2idx))
print(word2idx)
```

```python
vocab_size = len(word2idx) + 1        # Adding 1 to the vocab_size because the index starts from 1 not 0. This will make it uniform when using it further
print(vocab_size)
```

```python
input_seq = []

for line in final_data:
  token_list = tokenizer.texts_to_sequences([line])[0]
  for i in range(1, len(token_list)):
    n_gram_seq = token_list[:i+1]
    input_seq.append(n_gram_seq)

print(input_seq)
```

```python
max_seq_length = max(len(x) for x in input_seq)
print(max_seq_length)
```

```python
input_seq = np.array(pad_sequences(input_seq, maxlen=max_seq_length, padding='pre'))
print(input_seq)
```

```python
xs = input_seq[:, :-1]
labels = input_seq[:, -1]
print("xs: ",xs)
print("labels:",labels)
```

```python
from tensorflow.keras.utils import to_categorical
```

```python
ys = to_categorical(labels, num_classes=vocab_size)
print(ys)
```

```python
from tensorflow.keras.layers import Input, Dense, Embedding, LSTM, Dropout, Bidirectional, GlobalMaxPooling1D
from tensorflow.keras.models import Model
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.models import Sequential,Model
```

```python
 model = Sequential()
 model.add(Embedding(vocab_size, 124, input_length=max_seq_length-1))
 model.add(Dropout(0.2))
 model.add(LSTM(520, return_sequences=True))
 model.add(Bidirectional(LSTM(340, return_sequences=True)))
 model.add(GlobalMaxPooling1D())
 model.add(Dense(1024, activation='relu'))
 model.add(Dense(vocab_size, activation='softmax'))
```

```python
model.compile(optimizer='adam',
              loss = 'categorical_crossentropy',
              metrics=['accuracy'])
```

```python
model.summary()
```

```python
r = model.fit(xs,ys,epochs=100)
```

```python
import matplotlib.pyplot as plt
plt.plot(r.history['accuracy'])
```

```python
def predict_words(seed, no_words):
  for i in range(no_words):
    token_list = tokenizer.texts_to_sequences([seed])[0]
    token_list = pad_sequences([token_list], maxlen=max_seq_length-1, padding='pre')
    predicted = np.argmax(model.predict(token_list), axis=1)

    new_word = ''

    for word, index in tokenizer.word_index.items():
      if predicted == index:
        new_word = word
        break
    seed += " " + new_word
  print(seed)
```
